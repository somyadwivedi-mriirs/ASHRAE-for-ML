# 0. Overall Purpose: <a class="anchor" id="Code_Objective_0."></a>

The purpose of these Jupyter notebooks is to investigate how much energy a building will consume based on data from over 1,000 buildings in a three-year timeframe provided in the 2019 Kaggle competition [[Ref. 2]](#Ref_2) , "ASHRAE - Great Energy Predictor III," using different Machine Learning (ML) algorithms.  

The ML algorithms are used to make accurate models of metered building energy usage in: 
* chilled water, 
* electric, 
* hot water, and 
* steam.    

The accuracy of these models is evaluated based on the Root Mean Squared Logarithmic Error (RMSLE):

\begin{equation*}
\epsilon = \sqrt{\frac{1}{n} \sum_{i=1}^n (\log(p_i + 1) - \log(a_i+1))^2 }
\end{equation*}

Where:

* `Ïµ` is the RMSLE value (score) 
* `n` is the total number of observations in the (public/private) data set,
* `pi` is your prediction of target, and 
* `ai` is the actual target for `i`. 
* `log(x)` is the natural logarithm of `x`.

#### 13.  Wind Direction: `wind_direction`

* Wind Direction, the compass direction (0-360 degrees) the wind is blowing.
    * continuous: units = degrees

#### 14.  Wind Speed: `wind_speed`

* Wind Speed, the speed the wind is blowing.

#### 15. Row ID: `row_id`

* Row ID, the row ID in the test.csv data file.
    * continuous: units = #  

#### 16.  Meter: `meter`

#### 17.  Meter Reading: `meter_reading`

```python
def file_loader(file_name,currentDirectory,default_path): 
    try:
        INPUT_path_default = os.path.join(default_path,file_name)
        data= pd.read_csv(INPUT_path_default)
    except OSError:
        INPUT_path=currentDirectory+default_path
        #print ("Input file not found at location:",INPUT_path_default)
        data_path=os.path.join(INPUT_path,file_name)
        data= pd.read_csv(data_path)
        print ("Successfully loaded Input file: ",file_name," from:",currentDirectory)
    else:
        print ("Successfully loaded Input file: ",file_name," from:",default_path)
    return data
```

## 1.4. Reduce Data Memory <a class="anchor" id="Code_Objective_1_4"></a>

This function comes from [CeasarLupum](https://www.kaggle.com/caesarlupum/ashrae-start-here-a-gentle-introduction) [[Ref. 11]](#Ref_11)
Please go and give him some upvotes! This is an excellent function to reduce memory for our problem.

#### Function: reduce_mem_usage

```python
## Function to reduce the DF size
def reduce_mem_usage(df, verbose=True):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2    
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)    
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose: print('Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)'.format(end_mem, 100 * (start_mem - end_mem) / start_mem))
    return df
```

```python
# Reduce Memory
building = reduce_mem_usage(building)
weather_train_data = reduce_mem_usage(weather_train_data)
weather_test_data = reduce_mem_usage(weather_test_data)
data_train = reduce_mem_usage(data_train)
data_test = reduce_mem_usage(data_test)
```
